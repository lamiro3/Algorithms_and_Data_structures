```python
#[11.3]

set_size = 0
parent = []

def init_set(nSets):
    global set_size, parent
    set_size = nSets
    for i in range(nSets):
        parent.append(-1)
        
def find(id):
    while (parent[id] >= 0):
        id = parent[id]
    return id # 최상위 부모노드 id 반환

def union(s1, s2):
    global set_size
    parent[s1] = s2
    set_size -= 1

def MSTKruskal(vertex, adj):
    vsize = len(vertex)
    init_set(vsize)
    elist = []
    
    for i in range(vsize-1):
        for j in range(i+1, vsize):
            if adj[i][j] != None:
                elist.append((i, j, adj[i][j]))
    
    elist.sort(key=lambda x: x[2], reverse=True) # 간선 리스트를 가중치의 내림차순으로 정렬

    edgeAccepted = 0
    
    while (edgeAccepted < vsize - 1):
        e = elist.pop()
        uset = find(e[0])
        vset = find(e[1])
        
        if uset != vset:
            print(f'간선 추가 : ({vertex[e[0]]}, {vertex[e[1]]}, {e[2]})')
            union(uset, vset)
            edgeAccepted += 1
            
vertex = ["A", "B", "C", "D", "E"]
adj = [
    [None, 1, None, 2, None],
    [1, None, 3, 1, None],
    [None, 3, None, 3, 1],
    [2, 1, 3, None, 2],
    [None, None, 1, 2, None]
]

print("MST By Kruskal's Algorithm")
MSTKruskal(vertex, adj)
```

```python
#[11.4]

INF = 1e5

def getMinVertex(dist, selected):
    minv = 0
    mindist = INF
    
    for v in range(len(dist)):
        if not selected[v] and dist[v] < mindist:
            mindist = dist[v]
            minv = v
    return minv

def MSTPrim(vertex, adj):
    vsize = len(vertex)
    dist = [INF]*vsize
    selected = [False]*vsize
    dist[0] = 0
    
    for i in range(vsize):
        u = getMinVertex(dist, selected)
        selected[u] = True
        print(vertex[u], end=' ')
        
        for v in range(vsize):
            if (adj[i][v] is not None):
                if not selected[v] and adj[u][v] < dist[v]:
                    dist[v] = adj[u][v]
    print()
    
vertex = ["A", "B", "C", "D", "E"]
adj = [
    [None, 1, None, 2, None],
    [1, None, 3, 1, None],
    [None, 3, None, 3, 1],
    [2, 1, 3, None, 2],
    [None, None, 1, 2, None]
]

print("MST By Prim's Algorithm")
MSTPrim(vertex, adj)
```

```python
#[11.10]

INF = 1e5

def choose_vertex(dist, found):
    min = INF
    minpos = -1
    for i in range(len(dist)):
        if dist[i]<min and not found[i]:
            min = dist[i]
            minpos = i
    return minpos

def Dijkstra(vtx, adj, start):
    vsize = len(vtx)
    dist = list(adj[start])
    path = [start]*vsize
    found = [False]*vsize
    found[start] = True
    dist[start] = 0
    
    for i in range(vsize):
        print("step%2d: "%(i+1), dist)
        u = choose_vertex(dist, found)
        found[u] = True
        
        for w in range(vsize):
            if not found[w]:
                if dist[u] + adj[u][w] < dist[w]:
                    dist[w] = dist[u] + adj[u][w]
                    path[w] = u
    return path

vertex = ["A", "B", "C", "D", "E", "S"]
adj = [
    [0, INF, INF, INF, 9, 14],
    [INF, 0, 10, 15, INF, 7],
    [INF, 10, 0, 11, INF, 9],
    [INF, 15, 11, 0, 6, INF],
    [9, INF, INF, 6, 0, INF],
    [14, 7, 9, INF, INF, 0]
]

print("Shortest Path By Dijkstra Algorithm")

# 시작 정점은 5 -> "S"로 선택
path = Dijkstra(vertex, adj, 5)

for end in range(len(vertex)):
    if end==4:
        print("[최단경로: %s->%s] %s" %
              (vertex[5], vertex[end], vertex[end]), end='')
        while path[end] != 5:
            print(" <- %s" %vertex[path[end]], end='')
            end = path[end]
        print(" <- %s" %vertex[path[end]])
```
