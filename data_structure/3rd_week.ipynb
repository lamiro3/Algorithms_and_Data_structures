```python
# P3-1
class ArrayList:
  def __init__(self):
    self.arr = []

  def size(self):
    return len(self.arr)

  def getEntry(self, i):
    return self.arr[i]
  
  def insert(self, pos, s):
    left = self.arr[:pos]
    right = self.arr[pos+1:]

    left.append(s)
    self.arr = left + right
  
  def delete(self, pos):
    temp = []
    while len(self.arr) > pos+1:
      temp.append(self.arr.pop(-1))
    
    self.arr.pop(-1)
    
    temp.reverse()
    self.arr =  self.arr + temp
  
  def replace(self, pos, s):
    self.arr[pos] = s
  
  def find(self, s):
    for i, a in enumerate(self.arr):
      if a == s:
        return i

  def merge(self, others):
    for other in others:
      self.insert(len(self.arr), other)

arr = ArrayList()

for _ in range(5):
  arr.insert(_, _+1)
print(arr.arr, " - 리스트에 원소 추가")

arr.delete(2)
print(arr.arr, " - 리스트에서 원소 제거")

print(arr.find(5), " - 해당 원소가 있는 위치 검색")

arr.merge([2, 6, 7, 9])
print(arr.arr, "- extend() ~ merge()으로 구현")
```

```python
#P3-2

def myLineEditor():
  List = ArrayList()
  while True:
    cmd = input("[메뉴선택] i-입력, d-삭제, r-변경, p-출력, l-파일읽기, s-저장, q-종료, f-탐색후출력=>")
    if cmd == 'i':
      pos = int(input(" 입력행 번호:"))
      Str = input(" 입력행 내용: ")
      List.insert(pos, Str)
    
    elif cmd == 'd':
      pos = int(input(" 삭제행 번호:"))
      List.delete(pos)
    
    elif cmd == 'r':
      pos = int(input(" 변경행 번호:"))
      Str = input(" 변경행 내용: ")
      List.replace(pos, Str)
    
    elif cmd == 'p':
      print('Line Editor')
      for line in range(List.size()):
        print("[%2d] " %line, end='')
        print(List.getEntry(line))

      print()
  
    elif cmd == 'l':
      tgtFN = input("파일명: ")
      infile = open(tgtFN, "r")
      lines = infile.readlines()

      for line in lines:
        List.insert(List.size(), line.rstrip('\n')) # 맨 뒤에 새 내용 추가
      infile.close()

    elif cmd == 's':
      tgtFN = input("파일명: ")
      outfile = open(tgtFN, "w")

      for i in range(List.size()):
        outfile.write(List.getEntry(i)+"\n")
      outfile.close()

    elif cmd == 'f':
      tgtFN = input("파일명: ")
      tgtfile = open(tgtFN, "r")
      lines = tgtfile.readlines()
      search = input("찾을 내용: ")
      
      print('Line Editor')
      for i, line in enumerate(lines):
        if search in line:
          print("[%2d] " %i, end='')
          print(line)

    elif cmd == 'q':
      break

myLineEditor()
```

```python
#P3-4
from copy import deepcopy

class read_poly():
  def __init__(self, new=[]):
    self.poly = []

    if len(new) == 0:
      N = int(input("다항식의 최고 차수를 입력하시오: "))
      for n in range(N, -1, -1):
        self.poly.append(int(input(f'x^{n}의 계수 : ')))

      self.poly.reverse() # 방법 2로 하기 위해 순서 전환
    
    else:
      self.poly = new

  def degree(self):
    return len(self.poly) - 1

  def Polynomial(self, new):
    return read_poly(new)
  
  def evaluate(self, scalar): #scalar 대입했을 때의 함숫값 반환환
    sum = 0

    for i, n in enumerate(self.poly):
      sum += (scalar**i)*n
    
    return sum
  
  def add(self, rhs):
    gap = len(self.poly) - len(rhs.poly)
    new = deepcopy(self.poly) if gap >= 0 else deepcopy(rhs.poly)
    
    for i in range(len(rhs.poly) if gap >=0 else len(self.poly)): # self가 더 길이가 길면 rhs 길이 만큼만 더해
      if gap >= 0:
        new[i] += rhs.poly[i]
      else:
        new[i] += self.poly[i]

    return self.Polynomial(new)

  def subtract(self, rhs):
    gap = len(self.poly) - len(rhs.poly)
    new = deepcopy(self.poly) if gap >= 0 else deepcopy(rhs.poly)
    
    for i in range(len(rhs.poly) if gap >=0 else len(self.poly)): 
      if gap >= 0:
        new[i] -= rhs.poly[i]
      else:
        new[i] -= self.poly[i]
        new[i] *= -1
    
    if gap < 0:
      for j in range(abs(gap)):
          new[-1-j] *= -1

    return self.Polynomial(new)

  def multiply(self, rhs):
    new = [0 for _ in range(len(self.poly)+len(rhs.poly) - 1)]
    for i, s in enumerate(self.poly):
      for j, r in enumerate(rhs.poly):
        new[i+j] += s*r
    return self.Polynomial(new)

  def display(self, left):
    result = left
    for i in range(len(self.poly) - 1, 0, -1):
      result += f'{self.poly[i]} x^{i} + '
    result += str(self.poly[0])

    print(result)

a = read_poly()
b = read_poly()
a.display("A(x) = ")
b.display("B(x) = ")

#덧셈
c = a.add(b)
c.display("<덧셈>C(x) = ")

#뺄셈
d = a.subtract(b)
d.display("<뺄셈>D(x) = ")

#곱셈
e = a.multiply(b)
e.display("<곱셈>E(x) = ")

#함숫값
print(f'<함숫값>A(3) = {a.evaluate(3)}')
```

```python
#[P3.3]

class Set:
  def __init__(self, items):
    self.items = items

  def __sub__(self, other):
    return self.difference(other)

  def contains(self, tgt):
    isC = True
    for i in range(len(self.items)):
      if self.items[i] == tgt:
        isC = True
        break
      else:
        isC = False
    return isC
  
  def delete(self, tgt):
    if self.contains(tgt):
      i = self.items.index(tgt)
      left = self.items[:i]
      right = self.items[i+1:]

      self.items = left + right
  
  def insert(self,tgt):
    if not self.contains(tgt):
      self.items.append(tgt)
      self.items.sort()

  def union(self, other): # 합집합
    new = Set([])
    for s in self.items:
      new.items.append(s)

    for o in other.items:
      if not self.contains(o):
        new.items.append(o)

    new.items.sort()
    return new.items
  
  def intersect(self, other): #교집합
    new = Set([])

    for o in other.items:
      if self.contains(o):
        new.items.append(o)

    new.items.sort()
    return new.items

  def difference(self, other): #차집합
    new = Set([])
    for s in self.items:
      new.items.append(s)

    for o in other.items:
      if new.contains(o):
        new.delete(o)
      else:
        continue
    
    new.items.sort()
    return new.items
    
s, a = Set([1, 2, 3, 4, 5, 6]), Set([4, 5, 6, 7, 8])

print(s.contains(4))

s.delete(4)
print(s.items)

s.insert(4)
print(s.items)

print(s.union(a))
print(s.intersect(a))
print(s-a)
```
